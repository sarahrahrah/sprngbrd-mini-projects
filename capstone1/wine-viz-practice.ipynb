```python
import numpy as np
import pandas as pd
```

```python
winedata = pd.read_csv("winedata.csv")
winedata.columns
```

```python
winedata.head(50)
```

```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')

sns.lmplot(x= 'points', y='price', data=winedata)
plt.show()
```

```python
#sns.lmplot(x= 'points', y='price', data=winedata, hue='country', palette='Set1')

```

```python
#besides linear regression, can do polynomial/higher-order regression in seaborn

plt.scatter(winedata['points'], winedata['price'], label='specific wine', color='maroon', marker='o')
sns.regplot(x='points', y='price', data=winedata, color='pink', scatter=None, order = 1, label='order 1')
sns.regplot(x='points', y='price', data=winedata, color='white', scatter=None, order = 2, label='order 2')
plt.legend(loc= 'upper right')
plt.show()







```

```python
sns.stripplot(x='country', y='price', data=winedata)
plt.title('Wines by Country')
plt.gcf().set_size_inches(30, 10)
plt.show()
```

```python
francewine = winedata[winedata['country'] == 'France']
francewine.info
```

```python
sns.lmplot(x= 'points', y='price', data=francewine)
plt.title('French Wines')
plt.show()
```

```python
sns.stripplot(x='province', y='price', data=francewine)
plt.title('French Wines by Region')
plt.gcf().set_size_inches(20, 5)
plt.show()
```

```python
uswine = winedata[winedata['country'] == 'US']
sns.lmplot(x= 'points', y='price', data=francewine)
plt.title('American Wines')
plt.show()
```

```python
sns.stripplot(x='province', y='price', data=uswine, size=4, jitter=True)
plt.title('American Wines by Region')
plt.gcf().set_size_inches(20, 5)
plt.show()
```

```python
calwine = uswine[uswine['province'] == 'California']
sns.lmplot(x= 'points', y='price', data=uswine)
plt.title('California Wines')
plt.show()
```

```python
calwine.head
```

```python
sns.stripplot(x='region_2', y='price', data=calwine, size=4, jitter=True)
plt.title('California Wines by Region')
plt.gcf().set_size_inches(20, 5)
plt.show()
```

```python
sns.violinplot(x='region_2', y='price', data=calwine)
plt.title('California Wines by Region')
plt.ylim([0,200])
plt.gcf().set_size_inches(10, 5)
plt.show()
```

```python
sns.violinplot(x='region_2', y='price', data=calwine, inner=None, color='lightgray')
sns.stripplot(x='region_2', y='price', data=calwine, size=1.5, jitter=True)
plt.title('California Wines by Region')
plt.ylim([0,200])
plt.gcf().set_size_inches(10, 5)
plt.show()
```

```python
sns.pairplot(calwine, kind = 'reg', hue='region_2')
```
